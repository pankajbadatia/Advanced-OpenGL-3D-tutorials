```
dataset_path = "/content/dataset/data/images"
new_dataset_path = "/content/dataset/new_data"
image_path = "/content/dataset/new_data/image"
gt_path = "/content/dataset/new_data/ground_truth"
```

```
import glob
files = [f for f in glob.glob(dataset_path + "**/*.jpg", recursive=True)]
```

```
from sklearn.datasets import load_sample_image
from sklearn.feature_extraction.image import extract_patches_2d
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from numpy import asarray
```

```
# !rm -rf  /content/dataset/new_data/image/
# !rm -rf  /content/dataset/new_data/image/
# !rm -rf  /content/dataset
```

```
name = 0
for each_image in files:
  name = name + 1
  print(each_image)
  # new_each_img_path = os.path.join(image_path,each_image)
  # new_each_gt_path = os.path.join(gt_path,each_image)
  img = Image.open(each_image).convert('L')
  data = asarray(img)
  # print('data shape = ',data.shape)
  plot_width, plot_height = data.shape
  # image = Image.fromarray(data)
  dpi = 100
  n_patch = 2
  width_inches, height_inches = plot_width/dpi, plot_height/dpi

  patches = extract_patches_2d(data, patch_size=(256, 256), max_patches = n_patch)

  for i, patch in enumerate(patches):
      im = Image.fromarray(patch, 'L')
      filename = str(name) + "_" +  str(i) + ".jpg"
      save_img_path = os.path.join(image_path,filename)
      print('save img path =',save_img_path)
      im.save(save_img_path)
```
